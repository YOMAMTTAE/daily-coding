```python
import pandas as pd

def calculate_pay(name, hours):
  wage = 20000
  lunch = 10000
  if hours > 8:
    pay = (8*wage)+((hours -8)*wage*1.5)
  else:
    pay = hours*wage
  return [name, hours, pay+lunch]

total_data = []
print("--- [현장 관리 시스템 v1.0] ---")

count = int(input("오늘 투입된 인원은 총 몇 명입니까? :"))

for i in range(count):
  print(f"\n{i+1}번째 인원 정보를 입력하세요.")
  name = input("성함:")
  hours = float(input("근무 시간: "))

  result = calculate_pay(name, hours)
  total_data.append(result)

df=pd.DataFrame(total_data, columns=["성함", "근무시간", "총정산액"])
df.to_csv("weekly_report.csv", index=False, encoding="utf-8-sig")

print("\n--- 정산 완료---")
print(df)
print("\n'weekly_report.csv' 파일이 생성되었습니다. 다운로드 하세요!")
```
