```python
def 정산기(이름, 시간):
    기본시급 = 20000
    식대 = 10000

    if 시간 > 8:
        연장시간 = 시간 - 8
        급여 = (8 * 기본시급) + (연장시간 * 기본시급 * 1.5)
    else:
        급여 = 시간 * 기본시급

    # [중요!] 여기서부터는 if/else 칸막이 밖입니다.
    # '급여' 상자는 위에서 이미 만들어졌으니, 누구든 여기서 식대를 더합니다.
    최종금액 = 급여 + 식대
    return f"{이름} 님의 최종 지급액은 {최종금액}원 입니다."

# 함수 밖 (맨 왼쪽 벽)
print(정산기("강민수", 8))
print(정산기("이진철", 10))
print(정산기("박기태", 6))
```
