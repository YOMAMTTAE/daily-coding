```python
import pandas as pd

data = [
    ["강민수", 8, 170000],
    ["이진철", 10, 230000],
    ["박기태", 6, 130000]
]

df = pd.DataFrame(data, columns=["성함", "근무시간", "최종정산액"])

print("--- 생성된 데이터프레임 확인 ---")
print(df)

df.to_csv("오늘의_현장정산.csv", index=False, encoding="utf-8-sig")

print("\n파일 저장 완료! 왼쪽 폴더 아이콘에서 '오늘의_현장정산.csv'를 찾아보세요.")
```
